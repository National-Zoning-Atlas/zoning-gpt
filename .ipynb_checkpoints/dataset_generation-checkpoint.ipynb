```python
def convert_non_float(x):
    if isinstance(x, str):
        try:
            return float(x)
        except:
            flt_lst = re.findall("\d+\.\d+", x)
            int_lst = re.findall(r'\d+', x)
            if flt_lst:
                return flt_lst[0]
            elif int_lst:
                return float(int_lst[0])
            else:
                return float('NaN')
    else:
        return x
```

```python
df['1-Family Min. Lot (ACRES) converted'] = df['1-Family Min. Lot (ACRES)'].apply(lambda x: convert_non_float(x))
df['2-Family Min. Lot (ACRES) converted'] = df['2-Family Min. Lot (ACRES)'].apply(lambda x: convert_non_float(x))
df['3-Family Min. Lot (ACRES) converted'] = df['3-Family Min. Lot (ACRES)'].apply(lambda x: convert_non_float(x))
df['4+-Family Min. Lot (ACRES) converted'] = df['4+-Family Min. Lot (ACRES)'].apply(lambda x: convert_non_float(x))
```

```python
df['State'] = 'CT'
```

```python
# rearrange columns
cols = list(df.columns.values)
cols = ['State'] + cols[:-1]
df = df[cols]
```

```python
df['Input'] = df['State'] + ', ' + df['Jurisdiction'] + ', ' + df['Full District Name']
```

```python
df = df.rename(columns={"1-Family Min. Lot (ACRES) converted": "1-Family-Min-Lot-Acres", 
                   "2-Family Min. Lot (ACRES) converted": "2-Family-Min-Lot-Acres", 
                   "3-Family Min. Lot (ACRES) converted": "3-Family-Min-Lot-Acres", 
                   "4+-Family Min. Lot (ACRES) converted": "4+-Family-Min-Lot-Acres"})
```
